```python
import matplotlib.pyplot as plt
```

```python
N = [100000,200000,250000,500000,750000,1000000]
```

## Processing Elements = 4 

```python
serialTime = ""
with open("serial.txt","r") as fs:
    serialTime=fs.read()
```

```python
serialTime = serialTime.split("\n")
```

```python
serialTime.pop()
```

```python
serialTime
```

```python
serialTime = [float(x) for x in serialTime]
```

```python
parallelTime4 = ""
with open("parallel_4.txt","r") as fs:
    parallelTime4 = fs.read()
parallelTime4 = parallelTime4.split("\n")
parallelTime4.pop()
parallelTime4 = [float(x) for x in parallelTime4]
```

```python
parallelTime4
```

```python
speedUp4 = []

for i in range(len(parallelTime4)):
    speedup = serialTime[i] / parallelTime4[i]
    speedUp4.append(speedup)
```

```python
speedUp4
```

```python
plt.plot(N, speedUp4, marker='o', color='b')
plt.xlabel('N (Number of Iterations)')
plt.ylabel('SpeedUp')
plt.title('SpeedUp for 4 Processing Elements vs N')
plt.grid(True)

plt.xticks(rotation=45)
plt.show()
```
